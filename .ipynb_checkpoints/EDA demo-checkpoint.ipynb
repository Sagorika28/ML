# Univariate,Bivariate and MultiVariate Analysis

```python
# below commands are not required if we're using pyforest
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# The %matplotlib inline command tells the IPython environment to draw the plots immediately after the current cell. The drawn plots are shown below the code and stored in the notebook document for future reference.
%matplotlib inline 
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

## Univariate Analysis

```python
df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
# using sepal length to classify the species
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal length')
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue="species", height=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
```

## Multivariate Analysis

```python
sns.pairplot(df, hue="species", height=3, palette="bright")
```
